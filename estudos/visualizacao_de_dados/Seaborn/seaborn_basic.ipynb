### Seaborn: Introdução

Seaborn é uma biblioteca de visualização de dados baseada no Matplotlib que fornece uma interface de alto nível para criar gráficos estatísticos bonitos e informativos. Ele facilita a criação de gráficos complexos com menos código e fornece opções adicionais para estilizar gráficos de maneira atraente.

Nesta seção, veremos exemplos básicos de:
- Gráficos de dispersão
- Gráficos de linha
- Gráficos de barras
- Histogramas
- Mapas de calor (Heatmaps)
- Gráficos de distribuição conjunta (Jointplots)

---

### Importação do Seaborn

#### 1. **Gráfico de Dispersão**

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Dados de exemplo
np.random.seed(0)
df = pd.DataFrame({
    'x': np.random.rand(100),
    'y': np.random.rand(100),
    'categoria': np.random.choice(['A', 'B'], size=100)
})

# Criar o gráfico de dispersão
sns.scatterplot(data=df, x='x', y='y', hue='categoria', style='categoria', palette='viridis')
plt.title('Gráfico de Dispersão com Categorias')
plt.show()
```

---
#### 2. Gráfico de Linha

Gráficos de linha são úteis para visualizar tendências temporais ou contínuas.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Gerar dados de exemplo
x = np.linspace(0, 10, 100)  
y = np.sin(x)                

# Garantir que os dados são arrays 1D (apenas precaução)
x = np.array(x).flatten()
y = np.array(y).flatten()

# Criar o gráfico de linha
sns.lineplot(x=x, y=y)
plt.title('Gráfico de Linha')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.show()
```

#### 3. Gráfico de Barras

Gráficos de barras ajudam a visualizar comparações entre categorias.

```python
# Dados de exemplo
df_bar = pd.DataFrame({
    'Categoria': ['A', 'B', 'C', 'D'],
    'Valores': [4, 7, 1, 8]
})

# Criar o gráfico de barras
sns.barplot(data=df_bar, x='Categoria', y='Valores', palette='coolwarm')
plt.title('Gráfico de Barras')
plt.show()
```

#### 4. Histograma

Histogramas mostram a distribuição de uma variável.

```python
# Dados de exemplo
dados = np.random.randn(1000)

# Criar o histograma
sns.histplot(dados, bins=30, kde=True, color='blue')
plt.title('Histograma com KDE')
plt.xlabel('Valores')
plt.ylabel('Frequência')
plt.show()
```

#### 5. Mapa de Calor (Heatmap)

Mapas de calor são úteis para visualizar correlações ou matrizes.

```python
# Dados de exemplo
matriz = np.corrcoef(np.random.randn(10, 10))

# Criar o mapa de calor
sns.heatmap(matriz, annot=True, fmt=".2f", cmap='coolwarm', square=True)
plt.title('Mapa de Calor (Correlação)')
plt.show()
```

#### 6. Gráfico de Distribuição Conjunta (Jointplot)

O Jointplot mostra a relação entre duas variáveis, bem como suas distribuições individuais.

```python
# Dados de exemplo
np.random.seed(0)
x = np.random.randn(100)
y = 2 * x + np.random.normal(0, 0.5, 100)

# Criar o gráfico de distribuição conjunta
sns.jointplot(x=x, y=y, kind='scatter', color='purple')
plt.suptitle('Gráfico de Distribuição Conjunta', y=1.02)
plt.show()
```
