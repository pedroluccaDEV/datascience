### Matplotlib: Introdução

Matplotlib é uma biblioteca Python para criação de gráficos estáticos, animados e interativos. É amplamente utilizada para visualização de dados em ciência de dados, aprendizado de máquina e análise estatística.

Nesta seção, veremos exemplos básicos de:
- Gráficos de linha
- Gráficos de dispersão (scatter)
- Gráficos de barras
- Histogramas
- Gráficos de pizza

---


#### 1. **Gráfico de Linha**

```python
import matplotlib.pyplot as plt
import numpy as np

# Dados de exemplo
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Criar o gráfico de linha
plt.plot(x, y, label='Seno')
plt.title('Gráfico de Linha')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.legend()
plt.grid(True)
plt.show()
```

#### 2. Gráfico de Dispersão (Scatter)

    Um gráfico de dispersão é usado para mostrar a relação entre duas variáveis.

```python
# Dados de exemplo
x = np.random.rand(50)
y = np.random.rand(50)
cores = np.random.rand(50)

# Criar o gráfico de dispersão
plt.scatter(x, y, c=cores, cmap='viridis', alpha=0.8)
plt.title('Gráfico de Dispersão')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.colorbar(label='Cor')
plt.show()
```

#### 3. Gráfico de Barras

    Gráficos de barras são usados para comparar categorias ou conjuntos de dados.

```python
# Dados de exemplo
categorias = ['A', 'B', 'C', 'D']
valores = [5, 7, 3, 8]

# Criar o gráfico de barras
plt.bar(categorias, valores, color='skyblue')
plt.title('Gráfico de Barras')
plt.xlabel('Categorias')
plt.ylabel('Valores')
plt.show()
```

#### 4. Histograma

    Um histograma é usado para representar a distribuição de frequência de um conjunto de dados.

```python
# Dados de exemplo
dados = np.random.randn(1000)

# Criar o histograma
plt.hist(dados, bins=30, color='purple', alpha=0.7)
plt.title('Histograma')
plt.xlabel('Valores')
plt.ylabel('Frequência')
plt.show()
```

#### 5. Gráfico de Pizza

    Gráficos de pizza são usados para mostrar proporções entre categoria

```python
# Dados de exemplo
labels = ['A', 'B', 'C', 'D']
valores = [15, 30, 45, 10]

# Criar o gráfico de pizza
plt.pie(valores, labels=labels, autopct='%1.1f%%', startangle=140, colors=['gold', 'lightblue', 'lightgreen', 'pink'])
plt.title('Gráfico de Pizza')
plt.axis('equal')  # Garantir que o gráfico seja circular
plt.show()
```
