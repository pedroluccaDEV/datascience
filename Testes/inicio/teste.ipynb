```python

# Importando bibliotecas
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Soma simples
a = 5
b = 10
soma = a + b
print(f"A soma de {a} e {b} é {soma}")
```

```python
# Gerando dados para um gráfico
x = np.linspace(0, 10, 100)
y = np.sin(x)
```

```python
# Plotando o gráfico
plt.plot(x, y)
plt.title("Gráfico de Seno")
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.grid(True)
plt.show()
```
